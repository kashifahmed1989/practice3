```python
def cal(num1,num2,sign):
    num1 = int(input('Enter a num: '))
    num2 = int(input('Enter a num: '))
    sign = input('Enter a sign: ')
    if sign == '+':
        return num1+num2
    elif sign == '-':
        return num1-num2
    elif sign == '*':
        return num1*num2
    else:
        return num1/num2
```

```python
cal('num1','num2','sign')
```

```python
print(cal(int,int,str))
```
